  <tr>

Asiático refere-se a asiáticos, ilhas do Pacífico e indianos. Nativos americanos / índios americanos estão incluídos na categoria "outros". Qualquer pessoa com etnia hispânica é classificada como hispânica, independentemente de sua raça registrada.</td>

  <tr>

    <td>Speeding (10–19 MPH Over Prima Facie Limit *)</td>

    <td>A violação cometida pelo motorista, padronizada em categorias que são consistentes entre os estados. </td>
    <td>Speeding</td>
  </tr>
  <tr>
    <td>search_conducted</td>
    <td>Um valor VERDADEIRO/FALSO indicando se uma pesquisa foi realizada. </td>
    <td>TRUE</td>
  </tr>
  <tr>
    <td>search_type_raw</td>
    <td>A justificativa para a pesquisa, na linguagem dos dados originais. NA, se nenhuma pesquisa foi realizada. Não há um formato padrão nos estados. Alguns estados têm várias justificativas para uma pesquisa. </td>
    <td>CONSENT SEARCH CONDUCTED
</td>
  </tr>

    <td>Consent</td>
  </tr>
  <tr>
    <td>contraband_found</td>

    <td>TRUE</td>
  </tr>
  <tr>
    <td>stop_outcome</td>
    <td>O resultado da parada. Muitos estados têm resultados caracteriticos - por exemplo, “CHP 215” na Califórnia - portanto, essa coluna não é padronizada entre os estados. "Citação" e "Aviso" são os valores que ocorrem mais comumente nos estados. Se a parada tiver múltiplos resultados, o desfecho mais grave é usado. Por exemplo, se uma parada resultasse em uma citação e um aviso, stop_outcome seria "Citação". </td>
    <td>Citation</td>

```python
import pandas as pd
import psycopg2 
import sqlalchemy
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
conn = psycopg2.connect('postgres://postgres:isis@localhost/california')
```

```python
cursor = conn.cursor()
data = pd.read_sql_query('select * from temporaria ORDER BY RANDOM() LIMIT 50000', conn)
data.to_csv('data.csv', index=False)
df = pd.read_csv('data.csv')
```

```python
df = pd.read_csv('data.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
#df.drop(columns=['id','stop_time', 'fine_grained_location','police_department', 'driver_age'], inplace=True)
```

```python
encoder = LabelEncoder()
df['state'] = encoder.fit_transform(df['state'])
df['ethnicity'] = encoder.fit_transform(df['ethnicity'])
df['is_arrested'] = encoder.fit_transform(df['is_arrested'])
df['search_conducted'] = encoder.fit_transform(df['search_conducted'])
df['driver_race'] = encoder.fit_transform(df['driver_race'])
df['driver_race_raw'] = encoder.fit_transform(df['driver_race_raw'])
```

```python
df.drop(columns='search_type', inplace=True)
df.info()
```

```python
n_df = df.dropna()
n_df.info()
```

```python
plt.figure(figsize=(10, 16))
sns.pairplot(n_df)
```

```python
n_df.hist(figsize=(15,15), bins=15)
```
